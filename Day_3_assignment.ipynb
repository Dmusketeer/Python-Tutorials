```python
# Que :1. Write a Python program to count the number of characters (character frequency) in a string:

string="Dheeraj Tiwari"
edict={}
for i in string:
    if i in edict:
        edict[i]+=1
    else:
        edict[i]=1
print('Number of Characters : ' ,edict)
```

```python
# Que :2. Write a Python program demonstrating all the string methods (that were discussed in day1 session).

s1='hey hOw aRe yoU !'
s2='hellO ! I\'m FINE'
s3 = "     banana     "
s4='98765432109876'
s5='1234'
print(type(s1))
print(type(s2))

print('length og s1 : ',len(s1))
print('length og s2 : ',len(s2))

print('convert s1 into upper case : ',s1.upper())
print('convert s2 into lower case : ',s1.lower())

print('convert s1 into upper case : ',s2.upper())
print('convert s2 into lower case : ',s2.lower())

print('convert 1st character into upper case ',s1.capitalize())
# print('convert 1st character into upper case ',s1.casefold())

print('cnter the string : ',s1.center(100))  #center align the string 
print('cnter the string : ',s1.center(100,'@'))  #center align the string with padding @

print('count num of h : ',s1.count('h')) #Returns the number of times a specified value occurs in a string
print('count num of whitespace : ',s2.count(' ')) #Returns the number of times a specified value occurs in a string

print('encode the string : ',s1.encode())  #Returns an encoded version of the string
print('encode the string : ',s2.encode())  #Returns an encoded version of the string

print('find the fiNe : ',s2.find('fiNe'))   #Searches the string for a specified value and returns the position of where it was found

print('My name is {} {}'.format('Dheeraj','Tiwari'))  #	Formats specified values in a string

print('banana'.ljust(20) ,'is my fevorite fruits' )  #Returns a left justified version of the string
print('banana'.rjust(20) ,'is my fevorite fruits' )  #Returns a right justified version of the string

print('lower case : ', s2.lower())    #Converts a string into lower case
 
print("Of All Fruits", s3.lstrip(), "is my favorite")  #Remove spaces to the left of the string
print("Of All Fruits", s3.rstrip(), "is my favorite")  #Remove spaces to the right of the string


print('split the string : ',s2[0:10].split())  #plits a string into a list

print('Swapcase : ',s2.swapcase())  # #Make the lower case letters upper case and the upper case letters lower case
print('Swapcase : ',s1.swapcase())  # #Make the lower case letters upper case and the upper case letters lower case

print('Swapcase : ',s2.title())   #Make the first letter in each word upper case
print('Swapcase : ',s1.title())   #Make the first letter in each word upper case

print('zfill : ',s5.zfill(10))    #Fill the string with zeros until it is 10 characters long
print('zfill : ',s4.zfill(10))    #value of the len parameter is less than the length of the string, no filling is done


```

```python
# Que :3. Write a Python program demonstrating all the list methods that were discussed in day1 session.


l1=[1,2,3,4,4,5,6,'hi']
l2=[2,2,6,7,7,8,10]
fruits = ['apple', 'banana', 'cherry', 'orange']

print('length of  the lists l1 : ',len(l1))  #length of the lists
print('length of  the lists l2 : ',len(l2))  #length of the lists

l1.append('Dheeraj')           #Adds an element at the end of the list    
print('append in l1 : ',l1)   

# l2.clear()                     #Removes all the elements from the list
# print('clear : ', l2)


l3=fruits.copy()                #Returns a copy of the list
print('copy of fruits list : ',l3)

print('count the num of elements : ',l1.count('hi'))  #	Returns a copy of the list
print('count the num of elements : ',l2.count(7))     #	Returns a copy of the list

l2.extend((10,11,12,13))         #Add the elements of a list (or any iterable), to the end of the current list
print('extend methods : ',l2)

print('index num of \'hi\' in l1 : ',l1.index('hi'))  #Returns the index of the first element with the specified value


l1.insert(3,'mom')#Adds an element at the specified position [list.insert(pos, elmnt)]
print('insert the elements at the specific position L1 : ',l1)

l1.pop(3)    #Removes the element at the specified position
print('Remove the element at specific position L1 : ',l1)
  
l1.remove('hi')  #Removes the first item with the specified value
print('remove the elements by its value : ',l1)

l1.reverse()  #Reverses the order of the list
print('reverse of the l1 list : ',l1)

l5=[12,455,678,232,678,90,3]
l5.sort(reverse=False)        #sorts the list ascending by default.
print('sort the list : ',l5)
l5.sort(reverse=True)          #sorts the list in descending order.
print('sort the list : ',l5)
```

```python
#  Que : 4. Write a Python program demonstrating all the dictionary methods that were discussed in day1 session.

md={
    "NAME":"DHEERAJ TIWARI",
    'AGE':'26',
    'DOB' :'12/1/1994' ,
    "brand": "Ford",
    "model": "Mustang",
    "year" : 1964
}
print('DICTIONARY : ',md)

print(100*"*")

# md.clear()  #Removes all the elements from the dictionary
# print('Clear the dict elements : ',md)

x=md.copy()     #Returns a copy of the dictionary
print('copy the elments of md into x : ',x)

print(100*"*")

x = ('key1', 'key2', 'key3')
y = {0:'hey',1:'how',2:'are'}
thisdict = dict.fromkeys(x, y)  #Returns a dictionary with the specified keys and value
print(thisdict)

print(100*"*")

print('Get the value  of the string : ',md.get('NAME'))  #Returns the value of the specified key

print(100*"*")

#Returns a list containing a tuple for each key value pair
print('item method to get the key value in the form of list of tuple : ',md.items())

print(100*"*")

print('Keys of dict : ', md.keys())  #Returns a list containing the dictionary's keys.

print(100*"*")

print('values of dict : ', md.values())  #Returns a list containing the dictionary's values.

print(100*"*")

md.pop('DOB')    #Removes the element with the specified key
print('remove the elements at specific key : ', md)

print(100*"*")


md.popitem() #Removes the last inserted key-value pair
print('pop items at last : ',md)

print(100*"*")

md.update(y) #Updates the dictionary with the specified key-value pairs
print('Updated dict : ', md)

```
